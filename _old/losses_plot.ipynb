```python
import json
import matplotlib.pyplot as plt
import os

# Get all directories in training_results
training_dirs = [d for d in os.listdir('training_results') if os.path.isdir(os.path.join('training_results', d))]

for training_dir in training_dirs:
    filename = os.path.join('training_results', training_dir, 'training_dynamics.json')
    print(training_dir)
    
    # Skip if training_dynamics.json doesn't exist
    if not os.path.exists(filename):
        continue
        
    # Load the losses
    with open(filename, 'r') as f:
        data = json.load(f)
    losses = data['losses']

    # Create minimalist plot
    plt.figure(figsize=(5, 3))
    plt.plot(losses, color='black', linewidth=1)
    plt.xlabel('Batch')
    plt.ylabel('Loss')
    plt.grid(True, alpha=0.3)
    plt.ylim(0, 1.0)
    plt.title(training_dir)
    plt.show()

    # Create minimalist plot with log x-axis
    plt.figure(figsize=(5, 3))
    plt.semilogx(losses, color='black', linewidth=1)
    plt.xlabel('Batch')
    plt.ylabel('Loss') 
    plt.grid(True, alpha=0.3)
    plt.xlim(100, len(losses)*1.1)
    plt.ylim(0, 1.0)
    plt.title(training_dir)
    plt.show()
```
