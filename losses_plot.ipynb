```python
import json
import matplotlib.pyplot as plt

filename = 'training_losses_subject2_trial4.json'
filename = 'training_losses_multisubject2.json'
filename = 'training_losses_multisubject_sampling_newmask.json'
filename = 'training_losses_multisubject_synth_mse.json'
# Load the losses
with open(filename, 'r') as f:
    data = json.load(f)
losses = data['losses']

# Create minimalist plot
plt.figure(figsize=(5, 3))
plt.plot(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)
plt.show()

# Create minimalist plot with log x-axis
plt.figure(figsize=(5, 3))
plt.semilogx(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss') 
plt.grid(True, alpha=0.3)
plt.show()

# Create minimalist plot with log x-axis
plt.figure(figsize=(5, 3))
plt.semilogx(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss') 
plt.grid(True, alpha=0.3)
#plt.xlim(100, 1300)
plt.ylim(-100, -40)
plt.show()

```

```python
# Create minimalist plot
plt.figure(figsize=(5, 3))
plt.plot(losses[:1336], color='black', linewidth=2, label='Subject 2, Trial 4')
plt.plot(losses[1336:2376], color='green', linewidth=2, label='Subject 1, Trial 1')
plt.plot(losses[2376:], color='red', linewidth=2, label='Subject 1, Trial 1')
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
