```python
import json
import matplotlib.pyplot as plt

# Load the losses
with open('training_losses_subject2_trial4.json', 'r') as f:
    data = json.load(f)
losses = data['losses']

# Create minimalist plot
plt.figure(figsize=(5, 3))
plt.plot(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)
plt.show()

# Create minimalist plot with log x-axis
plt.figure(figsize=(5, 3))
plt.semilogx(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss') 
plt.grid(True, alpha=0.3)
plt.show()

# Create minimalist plot with log x-axis
plt.figure(figsize=(5, 3))
plt.semilogx(losses, color='black', linewidth=2)
plt.xlabel('Batch')
plt.ylabel('Loss') 
plt.grid(True, alpha=0.3)
plt.xlim(100, 1300)
plt.ylim(-100, -40)
plt.show()

```
